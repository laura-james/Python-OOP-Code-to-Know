#A class defines methods and property/attribute fields that capture the common behaviours and characteristics of objects
## A class is basically a blueprint for the objects that are created from it

The example below shows a simple class called Animal (classes usually start with capital letters by convention)


Inside the class are usually a set of functions called ***methods***; in this case there are two methods:** __init__**() and makesound()

The **__init__()** method is a special method called a ***constructor***. It runs automatically the first time an object is created using the Class. You can think of it running when the object is *spawned*. The official term for this is ***instantiated***
(Notice all the methods have "self" as the first (or only) parameter in the brackets being passed to the function)

Inside the init method there are a set of special varaiables, called ***attributes*** that store the properties of the Class (in this case: name, numLegs and colour)

You can see on line 11 an object has been instantiated and given the identifier of mydog.

Line 12 runs the makesound() method

Line 13 prints out the value of the name attribute

(Notice Line 11, 12 and 13 aren't inside the class definition)

```python
class Animal():
  def __init__(self):
    self.name = "Bob"
    self.numLegs = 4
    self.colour = "brown"
    print("---spawned---")

  def makesound(self):
    print("woof")

mydog = Animal()
mydog.makesound()
print(mydog.name)
```

The example above is not that helpful as all objects created from that class will have the same properties - unlike Animals in the real world which each have different properties.

The following example adapts the code above to allow some attributes to be specified when the object is created.

```python
class Animal():
  def __init__(self, name, numLegs, colour):
    self.name = name
    self.numLegs = numLegs
    self.colour = colour
    print(name, "has been spawned!")
    
  def makesound(self):
    print("woof")

mydog = Animal("Clive", 4, "grey")
mydog.makesound()
print(mydog.name)

yourdog = Animal("Rocket", 4, "black and white")
yourdog.makesound()
print(yourdog.name)

print(type(mydog))
print(type(yourdog))
```

In the example above note that Clive and Rocket are two separate entities, even though they have been created with the same class (or blueprint) they are not linked. 

For example, if we change the colour of one, the other one isn't affected.

```python
yourdog.colour = "brown"
print(yourdog.colour)
print(mydog.colour)
```

Directly changing attributes as shown on line 1 above could have unintended consequences, for example changing numLegs to 5 would not make sense.
## Encapsulation
Encapsulation refers to the bundling of data with the mechanisms or methods that operate on the data. 

It also refers to the limiting of direct access to some of that data, such as an object's components. Encapsulation allows developers to present a consistent and usable interface which is independent of how a system is implemented internally

To prevent attributes being changed from outside the class, we can declare them to be **private** attributes. Currently the three attributes (name, colour, numLegs) are all **public**

```python
class Animal():
  def __init__(self, name, numLegs, colour):
    self.__name = name
    self.__numLegs = numLegs
    self.__colour = colour
    print(name, "has been spawned!")
    
  def makesound(self):
    print("woof")

yourdog = Animal("Rocket", 4, "black and white")
yourdog.makesound()
print(yourdog.__name)
```

Note: all private attributes in Python have double underscores at the start.

Line 13 which attempts to print the private attribute __name  now generates an error.

In order to retrieve (and update) these private attributes, and possibly prevent any unwanted changes, we need to add extra methods to cater for this. These are often called **getter** and **setter** methods.

```python
class Animal():
  def __init__(self, name, numLegs, colour):
    self.__name = name
    self.__numLegs = numLegs
    self.__colour = colour
    print(name, "has been spawned!")

  def getnumLegs(self):
    return self.__numLegs

  def setnumLegs(self, num):
    if num <= 4 and num >= 0: #between 0 and 4
      self.__numLegs = num
    else:
      self.__numLegs = 4 #default is 4

  def makesound(self):
    print("woof")

yourdog = Animal("Rocket", 4, "black and white")
yourdog.setnumLegs(3)
print("numLegs is", yourdog.getnumLegs())
yourdog.setnumLegs(99)
print("numLegs is", yourdog.getnumLegs())
```

The code above shows the setter and getter method for modifying and retrieving the numLegs private attribute. (I haven't added ones for the other two private attributes. You don't need to add them unless you need to get/set the attribute outside the object)

Methods can also be private and public and it follows the same ideas. Public methods can be run from outside the object and private methods can only be accessed inside the object. These are usually only called by other methods inside the object.

# Inheritance
When a class derives from another class the child class will inherit all the public (and protected) attributes and methods from the parent class. In addition, it can have its own properties and methods. 

```python
class Cat(Animal): # this is how you define a class to inherit from an existing class
  def __init__(self, name, numLegs, colour):
    self.__name = name
    self.__numLegs = numLegs
    self.__colour = colour
    print(name, " - a cat - has been spawned!")
  def makesound(self):
    print("Meow!")
  def purr(self):
    print("PRRRRRRRRRRRRRRRRRRR")

fatcat = Cat("chubby",4 , "black")
fatcat.makesound()
fatcat.purr()
```

In addition to it's own methods, a child class can run methods that are in it's parent class. See modified Animal and Cat class below:


```python
class Animal():
  def __init__(self, name, numLegs, colour):
    self.__name = name
    self.__numLegs = numLegs
    self.__colour = colour
    print(name, "has been spawned!")

  def getnumLegs(self):
    return self.__numLegs

  def setnumLegs(self, num):
    if num <=4:
      self.__numLegs = num
    else:
      self.__numLegs = 4 #default is 4

  def makesound(self):
    print("woof")

  def describe(self):
    print("I am a type of ",type(self))


class Cat(Animal):
  def __init__(self, name, numLegs, colour):
    self.__name = name
    self.__numLegs = numLegs
    self.__colour = colour
    print(name, " - a cat - has been spawned!")
  def makesound(self):
    print("Meow!")
  def purr(self):
    print("PRRRRRRRRRRRRRRRRRRR")

fatcat = Cat("chubby",4, "black")
fatcat.makesound()
fatcat.purr()
fatcat.describe()
yourdog = Animal("Rocket", 4, "black and white")
yourdog.describe()
```

# Using super()
We can make the code for the Cat class a little smaller by reusing the **__init__ ** method of the parent Animal class - as these are very similar.

To do this we need to use the special super() variable, which represents the parent class. 

Take a look at the reformatted code below - **the change is on line 26**

```python
class Animal():
  def __init__(self, name, numLegs, colour):
    self.__name = name
    self.__numLegs = numLegs
    self.__colour = colour
    print(name, "has been spawned!")

  def getnumLegs(self):
    return self.__numLegs

  def setnumLegs(self, num):
    if num <=4:
      self.__numLegs = num
    else:
      self.__numLegs = 4 #default is 4

  def makesound(self):
    print("woof")

  def describe(self):
    print("I am a type of ",type(self))


class Cat(Animal):
  def __init__(self, name, numLegs, colour):
    super().__init__(name, numLegs, colour) # this calls the parent class's __init__
    print(name, " - a cat - has been spawned!")
  def makesound(self):
    print("Meow!")
  def purr(self):
    print("PRRRRRRRRRRRRRRRRRRR")

fatcat = Cat("chubby",4, "black")
fatcat.makesound()
fatcat.purr()
fatcat.describe()
yourdog = Animal("Rocket", 4, "black and white")
yourdog.describe()
```

The reason for this that the __numLegs attribute is private and only accessible from inside the Animal object - and if we are using a Cat object (fatcat) it doesn't have access to this.

If we do want to allow access to parents' attributes from their child classes, we need to make the attributes ***protected***.

To do that in Python we name them with a single underscore and in a UML diagram these are marked with a # symbol.

Methods can also be public, protected and private as well, using the same naming conventions of no/single/double underscores as with the attributes  - the examples in this notebook have only dealt with public methods for ease.



```python
#ATTRIBUTES CHANGED TO PROTECTED now
class Animal():
  def __init__(self, name, numLegs, colour):
    self._name = name
    self._numLegs = numLegs
    self._colour = colour
    print(name, "has been spawned!")

  def getnumLegs(self):
    return self._numLegs

  def setnumLegs(self, num):
    if num <=4:
      self._numLegs = num
    else:
      self._numLegs = 4 #default is 4

  def makesound(self):
    print("woof")

  def describe(self):
    print("I am a type of ",type(self))


class Cat(Animal):
  def __init__(self, name, numLegs, colour):
    super().__init__(name, numLegs, colour) 
    print(name, " - a cat - has been spawned!")
  def makesound(self):
    print("Meow!")
  def purr(self):
    print("PRRRRRRRRRRRRRRRRRRR")

fatcat = Cat("chubby",4, "black")
fatcat.makesound()
fatcat.purr()
fatcat.describe()
yourdog = Animal("Rocket", 4, "black and white")
yourdog.describe()
fatcat._numLegs #should be able to access the numLegs attribute in parent as it's protected now
```

# Class / Static Variables
When we declare a variable inside a class but outside any method, it is called as class or static variable in python. Class or static variables can be referred through a class but not directly through an instance.

Class or static variables are quite distinct from and does not conflict with any other member variable with the same name. 

```python
#ATTRIBUTES CHANGED TO PROTECTED now
class Animal():
  count = 0 # initialise a ststic variable
  def __init__(self, name, numLegs, colour):
    self._name = name
    self._numLegs = numLegs
    self._colour = colour
    print(name, "has been spawned!")
    # static variables are referred to by their class name rather than teh object name
    # this is incrementing the "global" animal count each time an Animal is created
    # whenever this contructor method runs
    Animal.count = Animal.count + 1

  def getnumLegs(self):
    return self._numLegs

  def setnumLegs(self, num):
    if num <=4:
      self._numLegs = num
    else:
      self._numLegs = 4 #default is 4

  def makesound(self):
    print("woof")

  def describe(self):
    print("I am a type of ",type(self))


class Cat(Animal):
  def __init__(self, name, numLegs, colour):
    super().__init__(name, numLegs, colour) 
    print(name, " - a cat - has been spawned!")
  def makesound(self):
    print("Meow!")
  def purr(self):
    print("PRRRRRRRRRRRRRRRRRRR")

fatcat = Cat("chubby",4, "black")
yourdog = Animal("Rocket", 4, "black and white")
mydog = Animal("Clive", 4, "brown")
print(Animal.count)
```

# Other types of linking classes - Association (Composition and Aggregation)
While ineritance is generally thought of as an "is a" relationship (eg a Cat is an Animal) there are other types of relationships between classes.

## Association is a "has a" relationship
There are two types of Association

## Composition
If the main class is destroyed, the associated class would be destroyed as well?
Eg a Cat has a Tail

You can see from the example below that if the myCat variable has been destroyed, the Tail will be destroyed also.

(Also note how you can use the dot notation on the Tail object to print out it's attributes (they are public in this example, which might not be ideal in reality)

```python
class Tail():
  def __init__(self, length, isFluffy):
    self.length = length
    self.isFluffy = isFluffy
    print("Tail created")

class Cat(Animal):
  def __init__(self, name, numLegs, colour):
    super().__init__(name, numLegs, colour) 
    print(name, " - a cat - has been spawned!")
    self.myTail = Tail(20,True)

  def makesound(self):
    print("Meow!")
  def purr(self):
    print("PRRRRRRRRRRRRRRRRRRR")

fatcat = Cat("chubby",4, "black")
fatcat.myTail.length
fatcat.myTail.isFluffy
```

## Aggregation Association
This is another type of association but in this case when the original class is destroyed the associated class still remains.

eg if a Course is destroyed the Students still exist.

The example below shows a new class called Zoo which has a list of animals associated with it.

```python
class Zoo():
  def __init__(self, animals):
    print("-- New Zoo created!--")
    self.animals = animals
    for an in animals:
      print(an._name) #print out each animal name when Zoo is created

myZoo = Zoo(catarray) #this Zoo will have the 10 cats in it

yourZoo = Zoo([mydog,yourdog,fatcat]) #another Zoo with 3 animals
print()
yourZoo.animals[0].makesound() #makes teh 1st animal in yourZoo make a sound...
```

In the example above two new Zoo objects were created that each were associated with some Animal objects. However if the Zoos were deleted, the Animals would still exist.

# "Encapsulate what varies"
Encapsulating anything which is **likely to change** saves time. Where there is likely to be changes, then there should be a class for it.

**Encapsulation** refers to objects bundling all of their own attributes and methods so that using them does not affect other objects. When it comes to designing an object-oriented program, requirements which are likely to change in the future should be encapsulated so that the change can be carried out simply when required.

Encapsulating what varies minimises the volume of code which may need altering in the future.

# "Favour composition over inheritance"
Composition and aggregation is seen as superior to inheritance.

Favour Composition Over Inheritance (sometimes called the "composite reuse principle") is the idea that classes should carry out polymorphism through composition, rather than inheritance. Composition is more flexible than inheritance, it represents a weaker relationship than that of inheritance.

Some examples in OOP feature objects composed of other objects where the particular characteristics can't be seen as inherited, in these cases it is sensible to use composition rather than inheritance.

# "Program to Interfaces, not Implementation"
The principle of programming to interfaces, not implementation allows unrelated classes to use similar methods.

Let's say we have classes for a dishwasher, a car and an oven. They are all unrelated but share similar methods. For example, all three might have an "openDoor" method. An interface is a collection of abstract procedures that can be implemented by unrelated classes. Classes which implement the procedures do so by implementing the interface which contains them.

When creating an interface, you don't need to know ***how*** objects will implement the procedures.

## Virtual / Abstract classes

In Python there aren't exactly Interfaces (that are accessible for A Level students anyway!) 

The best way to explain this concept is with a **virtual class**. These are sometimes called abstract classes. This is a class that on its own shouldn't be used to create objects, but it is used to be inherited from and the child classes override the method "stubs" that have been created in their own unique ways. See below. The Vehicle class is the virtual class, the Car class inherits from it and overrides the 3 methods in its own way.

```python
class Vehicle(): #Vehicle is a virtual class
  def start(self): #these methods don't do anything
    pass           #but define what inheriting classes must define
  def move(self):
    pass
  def openDoor(self):
    pass
  def stop(self):
    pass

class Car(Vehicle):
  def start(self): 
    print("This car is starting")           
  def move(self):
    print("This car is moving")
  def openDoor(self):
    print("This car is opening a door")
  def stop(self):
    print("This car is stopping")
  
```
